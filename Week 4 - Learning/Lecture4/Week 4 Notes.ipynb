## Nearest-Neighbor Classification

__Nearest-Neighbor Classification__: Algorithm that, given an input, chooses the class of the nearest data point to that input.

However, nearest neighbor can sometimes make weird decisions along the borderline. 

__$k$-nearest-neighbor classification__: Algorithm that, given an inpur, chooses the most common class out of the $k$ nearest data points to that input. (Kind of like "voting".) This can work very well for solving a variety of problems. 

We'll see many algorithms. They each have their trade-offs.
